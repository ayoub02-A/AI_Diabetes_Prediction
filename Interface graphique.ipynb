```python
# -*- coding: utf-8 -*-

# Form implementation generated from reading ui file 'view.ui'
#
# Created by: PyQt5 UI code generator 5.15.6
#
# WARNING: Any manual changes made to this file will be lost when pyuic5 is
# run again.  Do not edit this file unless you know what you are doing.





class Ui_Dialog(object):
    
    #pour structurer notre interface
    def setupUi(self, Dialog):
        self.model = tf.keras.models.load_model("mod70.h5")
        Dialog.setObjectName("Dialog")
        Dialog.resize(487, 520)
        
        self.lineEdit = QtWidgets.QLineEdit(Dialog)
        self.lineEdit.setGeometry(QtCore.QRect(290, 80, 113, 20))
        self.lineEdit.setObjectName("lineEdit")
        self.lineEdit.setPlaceholderText("si homme 0")
        
        self.lineEdit_2 = QtWidgets.QLineEdit(Dialog)
        self.lineEdit_2.setGeometry(QtCore.QRect(290, 110, 113, 20))
        self.lineEdit_2.setObjectName("lineEdit_2")
        self.lineEdit_2.setPlaceholderText("30-200 mg/dl")
        
        self.lineEdit_3 = QtWidgets.QLineEdit(Dialog)
        self.lineEdit_3.setGeometry(QtCore.QRect(290, 150, 113, 20))
        self.lineEdit_3.setObjectName("lineEdit_3")
        self.lineEdit_3.setPlaceholderText("35-220")
        
        self.lineEdit_4 = QtWidgets.QLineEdit(Dialog)
        self.lineEdit_4.setGeometry(QtCore.QRect(290, 190, 113, 20))
        self.lineEdit_4.setObjectName("lineEdit_4")
        self.lineEdit_4.setPlaceholderText("00-99mm")
        
        self.lineEdit_5 = QtWidgets.QLineEdit(Dialog)
        self.lineEdit_5.setGeometry(QtCore.QRect(290, 220, 113, 20))
        self.lineEdit_5.setObjectName("lineEdit_5")
        self.lineEdit_5.setPlaceholderText("00-840")
        
        self.lineEdit_6 = QtWidgets.QLineEdit(Dialog)
        self.lineEdit_6.setGeometry(QtCore.QRect(290, 250, 113, 20))
        self.lineEdit_6.setObjectName("lineEdit_6")
        self.lineEdit_6.setPlaceholderText("00-70")
        
        self.lineEdit_7 = QtWidgets.QLineEdit(Dialog)
        self.lineEdit_7.setGeometry(QtCore.QRect(290, 280, 113, 20))
        self.lineEdit_7.setObjectName("lineEdit_7")
        self.lineEdit_7.setPlaceholderText("0-3")
       
        self.lineEdit_8 = QtWidgets.QLineEdit(Dialog)
        self.lineEdit_8.setGeometry(QtCore.QRect(290, 310, 113, 20))
        self.lineEdit_8.setObjectName("lineEdit_8")
        self.lineEdit_8.setPlaceholderText("1-100")
        
        self.label = QtWidgets.QLabel(Dialog)
        self.label.setGeometry(QtCore.QRect(100, 80, 81, 16))
        self.label.setObjectName("label")
        
        self.pushButton_2 = QtWidgets.QPushButton(Dialog,clicked=self.clearall)
        self.pushButton_2.setGeometry(QtCore.QRect(250, 350, 75, 23))
        self.pushButton_2.setObjectName("pushButton_2")
        
        self.label_10 = QtWidgets.QLabel(Dialog)
        self.label_10.setGeometry(QtCore.QRect(160, 20, 181, 51))
        
        font = QtGui.QFont()
        font.setPointSize(16)
        self.label_10.setFont(font)
        self.label_10.setObjectName("label_10")
        
        self.label_4 = QtWidgets.QLabel(Dialog)
        self.label_4.setGeometry(QtCore.QRect(100, 190, 71, 16))
        self.label_4.setObjectName("label_4")
        
        self.label_2 = QtWidgets.QLabel(Dialog)
        self.label_2.setGeometry(QtCore.QRect(100, 110, 51, 21))
        self.label_2.setObjectName("label_2")
        
        self.label_7 = QtWidgets.QLabel(Dialog)
        self.label_7.setGeometry(QtCore.QRect(100, 280, 131, 21))
        self.label_7.setObjectName("label_7")
        
        self.pushButton = QtWidgets.QPushButton(Dialog,clicked=self.calculation)
        self.pushButton.setGeometry(QtCore.QRect(150, 350, 75, 23))
        self.pushButton.setObjectName("pushButton")
        
        self.label_11 = QtWidgets.QLabel(Dialog)
        self.label_11.setGeometry(QtCore.QRect(320, 550, 111, 16))
        self.label_11.setObjectName("label_11")
        
        
        self.label_9 = QtWidgets.QLabel(Dialog)
        self.label_9.setGeometry(QtCore.QRect(200, 390, 241, 41))
        self.label_9.setObjectName("label_9")
        
        self.label_99 = QtWidgets.QLabel(Dialog)
        self.label_99.setGeometry(QtCore.QRect(100, 390, 241, 41))
        self.label_99.setObjectName("label_99")
        
        self.label_90 = QtWidgets.QLabel(Dialog)
        self.label_90.setGeometry(QtCore.QRect(220, 420, 241, 41))
        self.label_90.setObjectName("label_90")
        
        self.label_990 = QtWidgets.QLabel(Dialog)
        self.label_990.setGeometry(QtCore.QRect(100, 420, 241, 41))
        self.label_990.setObjectName("label_990")
        
        self.label_5 = QtWidgets.QLabel(Dialog)
        self.label_5.setGeometry(QtCore.QRect(100, 220, 47, 13))
        self.label_5.setObjectName("label_5")
        
        self.label_6 = QtWidgets.QLabel(Dialog)
        self.label_6.setGeometry(QtCore.QRect(100, 250, 47, 13))
        self.label_6.setObjectName("label_6")
        
        self.label_8 = QtWidgets.QLabel(Dialog)
        self.label_8.setGeometry(QtCore.QRect(100, 310, 47, 13))
        self.label_8.setObjectName("label_8")
        
        self.label_3 = QtWidgets.QLabel(Dialog)
        self.label_3.setGeometry(QtCore.QRect(100, 150, 81, 16))
        self.label_3.setObjectName("label_3")
        
        self.label_12 = QtWidgets.QLabel(Dialog)
        self.label_12.setGeometry(QtCore.QRect(190, 490, 111, 16))
        self.label_12.setObjectName("label_12")

        self.retranslateUi(Dialog)
        QtCore.QMetaObject.connectSlotsByName(Dialog)

        
        
        
        
    def retranslateUi(self, Dialog):
        #nommage des labels
        _translate = QtCore.QCoreApplication.translate
        Dialog.setWindowTitle(_translate("Dialog", "Diabetes prediction"))
        self.label.setText(_translate("Dialog", "Pregnancies"))
        self.pushButton_2.setText(_translate("Dialog", "reset"))
        self.label_10.setText(_translate("Dialog", "ÙŽDiabetes predictor"))
        self.label_4.setText(_translate("Dialog", "SkinThickness"))
        self.label_2.setText(_translate("Dialog", "Glucose"))
        self.label_7.setText(_translate("Dialog", "Diabetes Pedigree Function"))
        self.pushButton.setText(_translate("Dialog", "OK"))
        #self.label_11.setText(_translate("Dialog", "By Ayoub Ouhdidou"))
        self.label_9.setText(_translate("Dialog", " "))
        self.label_9.setFont(QFont("Arial",9, weight=QFont.Bold))
        self.label_99.setText(_translate("Dialog", "Outcome with deep :"))
        self.label_90.setText(_translate("Dialog", " "))
        self.label_90.setFont(QFont("Arial",9, weight=QFont.Bold))
        self.label_990.setText(_translate("Dialog", "Outcome with machine :"))
        self.label_5.setText(_translate("Dialog", "Insulin"))
        self.label_6.setText(_translate("Dialog", "BMI"))
        self.label_8.setText(_translate("Dialog", "Age"))
        self.label_3.setText(_translate("Dialog", "BloodPressure"))
        self.label_12.setText(_translate("Dialog", "By Ayoub Ouhdidou"))
        
        
        
    def calculation(self):
        #pour la prediction
        #feature_names ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 
        # 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
        s       =self.model
        arr     =[self.lineEdit,self.lineEdit_2,self.lineEdit_3,self.lineEdit_4,self.lineEdit_5,self.lineEdit_6,self.lineEdit_7,self.lineEdit_8]
        print(self.lineEdit.text())
        print(self.lineEdit_2.text())
        print(self.lineEdit_3.text())
        print(self.lineEdit_4.text())
        print(self.lineEdit_5.text())
        print(self.lineEdit_6.text())
        print(self.lineEdit_7.text())
        print(self.lineEdit_8.text())
        df = pd.DataFrame(arr)
        s=self.model
        arr ,ee = self.empty_entry_validation(arr)
        if ee == 0:
            arr     =list(map(float,arr))
            re = self.range_validation(arr)
            if re == 0: 
                self.arr=arr
                arr     =np.array([np.array(arr)])
                output = self.check_input(arr)
                output1= self.check_input1(arr)
                
                print(output)
                print(output1)
                if output>0.5:
                    output = output*100
                    self.label_9.setText("Diabetic with %.2f %%" % (float(output)))
                else:
                    output = output*100
                    output = 100 - output
                    self.label_9.setText("Not Diabetic with %.2f %%" % (float(output)))
                if output1==1:
                    self.label_90.setText("Diabetic ")
                else:
                    self.label_90.setText("Not Diabetic ")
            else:
                self.showdialog("invalid values as input")
        else:
            self.showdialog("please first enter the values")
            
        self.label_9.setFont(QFont("Arial",9, weight=QFont.Bold))
        
        
    @staticmethod
    def check_input(data):
        #load du model de deep learning
        
        df=pd.DataFrame(data=data,index=[0])
        p = keras.models.load_model("mod70.h5")  #model deep
        op = p.predict(df) 
       # print(self.op)
        return op[0]
    
    @staticmethod        
    def check_input1(data):
        #load du model de machine learning
        
        df1=pd.DataFrame(data=data,index=[0])
        pt = pickle.load(open('TR.pickle', 'rb'))   #model machine
        op = pt.predict(df1)
        return op[0]  
                
        
    # pour tester les valeurs dans les inputes
    def range_validation(self,arr):
        range_errors = 0
        if arr[0]>25:
            range_errors+=1
            self.lineEdit.clear()
        
        if arr[2]>220.00 or arr[2]<30.00:
            range_errors+=1
            self.lineEdit_3.clear()
        
        if arr[1]>200.00 or arr[1]<24.00:
            range_errors+=1
            self.lineEdit_2.clear()
        
        if arr[3]>100.00 or arr[3]<0.00:
            range_errors+=1
            self.lineEdit_4.clear()
        
        if arr[4]>900.00 or arr[4]<0.00:
            range_errors+=1
            self.lineEdit_5.clear()
        
        if arr[5]>70.00 or arr[5]<0.00:
            range_errors+=1
            self.lineEdit_6.clear()
        
        if arr[6]>3.00 or arr[6]<0.00:
            range_errors+=1
            self.lineEdit_7.clear()
        
        if arr[7]>100.00 or arr[7]<1.00 :
            range_errors+=1
            self.lineEdit_8.clear()
        return range_errors
    
    
    #pour tester les inputes
    def empty_entry_validation(self,arr):
        l=[]
        empty=0
        for x in arr:
            l.append(x.text())
        if '' in l :
            empty=1
        for i in range(len(l)):
            if l[i]=='.':
                arr[i].clear()
                empty=1
        return l,empty
    
    
    #pour vider les inputes
    def clearall(self):
        self.lineEdit.clear()
        self.lineEdit_2.clear()
        self.lineEdit_3.clear()
        self.lineEdit_4.clear()
        self.lineEdit_5.clear()
        self.lineEdit_6.clear()
        self.lineEdit_7.clear()
        self.lineEdit_8.clear()
        self.label_9.setText(" ")
        self.label_9.setFont(QFont("Arial",9, weight=QFont.Bold))
        self.label_90.setText(" ")
        self.label_90.setFont(QFont("Arial",9, weight=QFont.Bold))
        self.arr=['','','','','','','','']

        
    #dialog d'erreur
    def showdialog(self,s):
        msg = QtWidgets.QMessageBox()
        msg.setIcon(QtWidgets.QMessageBox.Information)
        msg.setText("Message Alert!")
        msg.setInformativeText("Additional information")
        msg.setWindowTitle("Attention!")
        msg.setDetailedText(s)
        msg.setStandardButtons(QtWidgets.QMessageBox.Ok | QtWidgets.QMessageBox.Cancel)
        msg.exec_()


        
        
    
        
        
        
if __name__=="__main__":
    app = QtWidgets.QApplication(sys.argv)
    Dialog = QtWidgets.QDialog()
    ui = Ui_Dialog()
    ui.setupUi(Dialog)
    Dialog.show()
    sys.exit(app.exec_())
```
